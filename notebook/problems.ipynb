## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

# Simulation of dice 1
dice_1 = np.random.randint(1, 7, 1000)

# Simulation of dice_2
dice_2 = np.random.randint(1, 7, 1000)

# Sum of both dices
total_dices = dice_1 + dice_2

# Condition to check:
# 1. Greater than 7
# or
# 2. even number

# counters
counter_expected = 0

for i in total_dices:
  if (i >= 7) or (i % 2 == 0):
    counter_expected += 1

# task 3
print(f"The estimated probability that the total score is even or greater than 7: {(counter_expected / 1000):.2}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
# ---------- Notes ---------- 
# A box contains:
# 10 W balls
# 20 R balls
# 30 G balls

# 1. We take 5 balls from the box with replacement

# 2. We take the ball, observe what color it is and put it back into the box

# What we want: We want to know the probability of:

    # a. Take 3 white and 2 red.
    # b. Take all of the same color.

# Run the experiment 1000 times and calculate the above probabilities

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

# print(ball_box)
            
# TODO
lista_ball_box = list(ball_box.values())

# Number of experiments
experiments = 1000

# task 1
red_2_white_3 = 0

# task 2
all_same_color = 0

for i in range(experiments):
    sacadas = np.random.choice(lista_ball_box, 5)

    white_count = np.sum(sacadas == "White")
    red_count = np.sum(sacadas == 'Red')
    green_count = np.sum(sacadas == 'Green')

    if white_count == 3 and red_count == 2:
        red_2_white_3 += 1
        
    if white_count == 5 or red_count == 5 or green_count == 5:
        all_same_color += 1

# Calculate the probability
# red: 2
# white: 3
p_red_2_white_3 = red_2_white_3 / experiments
print(f"{p_red_2_white_3:.2f}%")

# all same color
p_all_same_color = all_same_color / experiments
print(f"{p_all_same_color:.2f}%")

```
